```python
def addLine(text, line):
    return text + line + '\n'

def addChannel(text, index, type, module, channel, delay, link, start, stop, name):
    return addLine(text, f'{index}\t{type}\t{module}\t{channel}\t{delay}\t{link}\t{start}\t{stop}\t{name}')

def addMultiChannel(text, indexFrom, type, module, channelFrom, channelTo, delay, link, start, stop, name):
    indexNo = indexFrom 
    for i in range(channelFrom, channelTo+1):
        text = addLine(text, f'{indexNo}\t{type}\t{module}\t{i}\t{delay}\t{link}\t{start}\t{stop}\t{name}')
        indexNo += 1
    return text

def addIndies(text, indexStart, module, channelFrom, channelTo, delay, link, start, stop, name):
    detIndex = 1
    typeIndex = indexStart
    for i in range(channelFrom, channelTo+1):
        if(i%2 == 0 and i != channelFrom):
            typeIndex += 1
            detIndex = 1
        text = addLine(text, f'{detIndex}\t{typeIndex}\t{module}\t{i}\t{delay}\t{link}\t{start}\t{stop}\t{name}{typeIndex}')
        detIndex += 1
    return text
    
```
